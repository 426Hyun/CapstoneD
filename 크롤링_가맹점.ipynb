```python
# 라이브러리 임포트
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
# 크롬드라이버 세팅
driver = webdriver.Chrome("C:/Users/hyun_desk/Project_final/chromedriver.exe")

c_adress=[]

for i in range(1, 8):
    # 사용자가 입력한 날짜 페이지로 이동.
    driver.get("https://www.purmeecard.com/public.do?request=merchantSelect&name=&si=02&curPage="+str(i))
    time.sleep(3)
    # 소스코드 가져오기
    source = driver.page_source

    s1 = BeautifulSoup(source, "html.parser")

    s_adress=s1.find_all("td", class_="table_list_left")
    c_adress.append(s_adress)
driver.close()
```

```python
len(c_adress)
```

```python
t_adress=[]
# 주소에 내용외 불필요한 태그 제거하기
for i in range(len(c_adress)):
    for j in c_adress[i]:
        t_adress.append(j.text.replace(' ','').split())
t_adress
```

```python
import pandas as pd
```

```python
# 주소정보 t_adress를 4개의 인덱스씩 출력
# 한 가맹점당 해당 정보
#f = codecs.open('D:/final_project/result.txt', encoding="utf-8", mode="w")
start_index=0
end_index=len(t_adress)
div = 4
adress_data=[]
for item in range(start_index, end_index+div, div):
    r_adress=t_adress[start_index:start_index+div]
    if r_adress != []:
        adress_data.append(r_adress)
        #f.write(r_adress + '\r\n')
    start_index = start_index+div
#f.close()
```

```python
# 크롤링 정보 데이터 프레임화
print(adress_data[0])
data=pd.DataFrame(adress_data)
data.columns = ['dong', 'store','tell','adr']
data.head()
```

```python
# csv파일로 저장
data.to_csv('result_adress.csv', encoding='cp949')
```
